## The dataset
We'll use **`monthly_pl.csv`** (included in your download). Two ways to load it:

**Option A — Upload to Colab:** run the cell below and pick the file.  
**Option B — Auto-generate:** if no file is found, the notebook builds the same data so you can run everything instantly.

```
import pandas as pd
import numpy as np

try:
    df = pd.read_csv("monthly_pl.csv", parse_dates=["Month"])
    print("Loaded monthly_pl.csv ✅")
except FileNotFoundError:
    # Fallback so the notebook always runs
    np.random.seed(42)
    months = pd.date_range("2023-01-01","2025-12-01",freq="MS")
    n=len(months); trend=np.linspace(1.0,1.6,n)
    seas=1+0.15*np.sin(np.arange(n)/12*2*np.pi)
    rev=(800000*trend*seas*(1+np.random.normal(0,0.05,n))).round(0)
    df=pd.DataFrame({"Month":months,"Revenue":rev,
        "COGS":(rev*np.random.uniform(.42,.48,n)).round(0),
        "OpEx":(rev*np.random.uniform(.28,.34,n)).round(0)})
    df["Gross_Profit"]=df.Revenue-df.COGS
    df["EBITDA"]=df.Gross_Profit-df.OpEx
    print("Generated data in-memory ✅")

df.head()
```

### 1. Get to know your data
Three commands you'll use every single time.

```
print("Shape (rows, cols):", df.shape)
print("\nColumn names:", list(df.columns))
df.describe().round(0)
```

### 2. Select columns and rows (like highlighting ranges)

```
# One column
print(df["Revenue"].head(3))

# Multiple columns
print(df[["Month","EBITDA"]].head(3))

# Filter rows: only months where EBITDA exceeded 150k
strong = df[df["EBITDA"] > 150_000]
print(f"\n{len(strong)} months had EBITDA over $150k")
```

### 3. Create new columns — vectorized formulas
No dragging a formula down 36 rows. One line does all of them.

```
df["Gross_Margin"]  = df["Gross_Profit"] / df["Revenue"]
df["EBITDA_Margin"] = df["EBITDA"] / df["Revenue"]
df[["Month","Revenue","Gross_Margin","EBITDA_Margin"]].head()
```

### 4. Sort and rank — find your best and worst months

```
best = df.sort_values("EBITDA", ascending=False).head(5)
print("Top 5 EBITDA months:")
best[["Month","Revenue","EBITDA","EBITDA_Margin"]]
```

### 5. Group and summarize by year (think PivotTable)

```
df["Year"] = df["Month"].dt.year
yearly = df.groupby("Year").agg(
    Revenue=("Revenue","sum"),
    EBITDA=("EBITDA","sum"),
    Avg_Margin=("EBITDA_Margin","mean")
).round(2)
yearly
```

### 6. Export back to Excel for your board pack

```
yearly.to_excel("yearly_summary.xlsx")
print("Saved yearly_summary.xlsx — find it in the Colab file panel (📁 on the left).")
```
