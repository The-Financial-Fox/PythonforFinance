### 1. Your very first line of Python
Think of `print()` as typing into a cell and hitting Enter.

```
print("Hello, I am a CFO who codes now.")
```

### 2. Variables — naming things
In Excel you might put `1,250,000` in cell B2 and call it *Revenue*. Same idea here.

```
revenue = 1_250_000      # underscores are just for readability
cogs    = 540_000
opex    = 310_000

gross_profit = revenue - cogs
ebitda       = gross_profit - opex

print("Gross Profit:", gross_profit)
print("EBITDA:", ebitda)
```

### 3. Margins — Python does the arithmetic
Notice how a formula is written once and reused.

```
gross_margin = gross_profit / revenue
ebitda_margin = ebitda / revenue

print(f"Gross Margin:  {gross_margin:.1%}")   # .1% formats as a percent
print(f"EBITDA Margin: {ebitda_margin:.1%}")
```

### 4. f-strings — building a clean report line
The `f` before the quotes lets you drop variables right into text. `:,.0f` means *comma thousands, no decimals*.

```
company = "Acme Corp"
print(f"{company} earned ${revenue:,.0f} in revenue and ${ebitda:,.0f} EBITDA.")
```

### 5. A reusable function — your first 'template'
A function is a saved formula you can call again and again with new inputs.

```
def ebitda_margin(revenue, cogs, opex):
    return (revenue - cogs - opex) / revenue

# Test it on three scenarios
print("Base case:    ", f"{ebitda_margin(1_250_000, 540_000, 310_000):.1%}")
print("Cost cut:     ", f"{ebitda_margin(1_250_000, 480_000, 280_000):.1%}")
print("Revenue boost:", f"{ebitda_margin(1_400_000, 600_000, 330_000):.1%}")
```
