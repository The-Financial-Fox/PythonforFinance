## Load the P&L (same loader as Notebook 2)

```
import pandas as pd, numpy as np
import matplotlib.pyplot as plt

try:
    df = pd.read_csv("monthly_pl.csv", parse_dates=["Month"])
except FileNotFoundError:
    np.random.seed(42)
    months=pd.date_range("2023-01-01","2025-12-01",freq="MS"); n=len(months)
    trend=np.linspace(1,1.6,n); seas=1+0.15*np.sin(np.arange(n)/12*2*np.pi)
    rev=(800000*trend*seas*(1+np.random.normal(0,0.05,n))).round(0)
    df=pd.DataFrame({"Month":months,"Revenue":rev,
       "COGS":(rev*np.random.uniform(.42,.48,n)).round(0),
       "OpEx":(rev*np.random.uniform(.28,.34,n)).round(0)})
    df["Gross_Profit"]=df.Revenue-df.COGS; df["EBITDA"]=df.Gross_Profit-df.OpEx

df["EBITDA_Margin"]=df.EBITDA/df.Revenue
df.head(3)
```

### 1. Revenue trend line — the CFO classic

```
plt.figure(figsize=(11,4))
plt.plot(df["Month"], df["Revenue"], linewidth=2.2, color="#1f4e79")
plt.title("Monthly Revenue Trend", fontsize=14, fontweight="bold")
plt.ylabel("Revenue ($)")
plt.grid(alpha=0.3)
plt.gca().yaxis.set_major_formatter(lambda x,_: f"${x/1e6:.1f}M")
plt.tight_layout(); plt.show()
```

### 2. Stacked cost structure — where the money goes

```
plt.figure(figsize=(11,4))
plt.stackplot(df["Month"], df["COGS"], df["OpEx"], df["EBITDA"],
              labels=["COGS","OpEx","EBITDA"],
              colors=["#c0504d","#f0ad4e","#4f81bd"], alpha=0.85)
plt.title("Revenue Composition Over Time", fontweight="bold")
plt.legend(loc="upper left"); plt.grid(alpha=0.3)
plt.tight_layout(); plt.show()
```

### 3. Margin trend with a target line

```
plt.figure(figsize=(11,4))
plt.plot(df["Month"], df["EBITDA_Margin"]*100, color="#4f81bd", linewidth=2)
plt.axhline(22, color="red", linestyle="--", label="Target 22%")
plt.title("EBITDA Margin vs Target", fontweight="bold")
plt.ylabel("Margin (%)"); plt.legend(); plt.grid(alpha=0.3)
plt.tight_layout(); plt.show()
```

### 4. A 4-panel KPI dashboard
Four charts in one figure — drop this straight into your deck.

```
df["Year"]=df["Month"].dt.year
yearly=df.groupby("Year")[["Revenue","EBITDA"]].sum()

fig, ax = plt.subplots(2,2, figsize=(13,8))
fig.suptitle("CFO KPI Dashboard", fontsize=16, fontweight="bold")

ax[0,0].plot(df["Month"], df["Revenue"], color="#1f4e79"); ax[0,0].set_title("Revenue Trend")
ax[0,1].bar(yearly.index, yearly["EBITDA"], color="#4f81bd"); ax[0,1].set_title("EBITDA by Year")
ax[1,0].plot(df["Month"], df["EBITDA_Margin"]*100, color="green"); ax[1,0].set_title("EBITDA Margin %")
ax[1,1].bar(yearly.index, yearly["Revenue"], color="#7f7f7f"); ax[1,1].set_title("Revenue by Year")
for a in ax.flat: a.grid(alpha=0.3)
plt.tight_layout(); plt.savefig("kpi_dashboard.png", dpi=150)
plt.show()
print("Saved kpi_dashboard.png ✅")
```
